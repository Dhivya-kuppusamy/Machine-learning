```python
from sklearn.model_selection import train_test_split
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.svm import SVC
from sklearn import svm
import numpy as np
```

```python
df1=pd.read_csv('https://raw.githubusercontent.com/nunnarilabs/ml/master/titanic/train.csv')
```

```python
df1.head()
```

```python
df1
```

```python
df1.drop(['PassengerId','Name','Ticket'],axis=1,inplace=True)
df1
```

```python
df1.isnull().sum()
```

```python
df1[df1.Embarked.isnull()]
```

```python
df1.drop(['Cabin'],axis=1,inplace=True)
df1["Age"]=df1["Age"].fillna(df1["Age"].median())
df1["Embarked"]=df1["Embarked"].fillna("C")
df1.Sex=df1.Sex.replace(['male','female'],[0,1])
df1.Embarked=df1.Embarked.replace(['S','C','Q'],[0,1,2])
df1
```

```python
df1.dropna(inplace=True)
```

```python
df2=pd.read_csv("https://raw.githubusercontent.com/nunnarilabs/ml/master/titanic/test.csv")
```

```python
df2.drop(['Cabin'],axis=1,inplace=True)
df2["Age"]=df2["Age"].fillna(df2["Age"].median())
df2["Embarked"]=df2["Embarked"].fillna("C")
df2.Sex=df2.Sex.replace(['male','female'],[0,1])
df2.Embarked=df2.Embarked.replace(['S','C','Q'],[0,1,2])
df2
```

```python
df2.dropna(inplace=True)
```

```python
features=["Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"]
```

```python
C=0.7
clf=svm.SVC(kernel='linear',C=C).fit(df1[features],df1["Survived"])
```

```python
prediction=clf.predict(df2[features])
```
