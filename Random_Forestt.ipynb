```python
data['Gender'] = data['Gender'].map({'Male' : 1,'Female' : 0})
data['Married'] = data['Married'].map({'Yes' : 1,'No' : 0})
data['Education'] = data['Education'].map({'Graduate' : 1,'Not Graduate' : 0})
data['Dependents'].replace('3+',inplace = True)
data['Loan_Status'] = data['Loan_Status'].map({'Y' : 1,'N' : 0})
data['Self_Employed'] = data['Self_Employed'].map({'Yes' : 1,'No' : 0})
data['Property_Area'] = data['Property_Area'].map({'Semiurban' : 1,'Urban' : 0,'Rural' : 1})
```

```python
for i in data.columns:
  data[i].fillna(data[i].mode()[0],inplace=True)
data
```

```python
data.isnull().sum()
```

```python
X=data.drop(columns=['Loan_ID','Loan_Status']).values
y=data['Loan_Status'].values
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sta = StandardScaler()
X_train=sta.fit_transform(X_train)
X_test=sta.transform(X_test)
```

```python
data.isnull().sum()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
rfc=RandomForestClassifier(criterion='entropy',random_state=42)
rfc.fit(X_train,y_train)
```

```python
data.isnull().sum()
```

```python
y_pred = rfc.predict(X_train)
print(classification_report(y_train,y_pred))

```

```python
y_pred_test = rfc.predict(X_test)
print(classification_report(y_test,y_pred_test))
```
