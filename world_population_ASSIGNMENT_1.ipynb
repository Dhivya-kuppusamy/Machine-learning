```python
from scipy import stats     # scipy for statistics eg.z-score
hp_zscore = np.abs(stats.zscore(data['2015-Population'])) #abs - ignoring the directions
hp_zscore
```

```python
thres = hp_zscore.max() #outlier as threshold
ind_val = np.where(hp_zscore>=thres)
data.iloc[ind_val]["2015-Population"]
```

```python
new_data= data.drop(index=data.iloc[ind_val]['2015-Population'].index) #drop the outlier
new_data
```
